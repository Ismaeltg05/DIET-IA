# Convertir JSON A CSV

```python
# 1. Cargar el archivo JSON
try:
    with open('full_format_recipes.json', 'r', encoding='utf-8') as f:
        data = json.load(f)

    # 2. Convertir a DataFrame de Pandas
    df = pd.DataFrame(data)

    # 3. Detectar columnas que son listas (ingredientes, pasos, categorias)
    # y convertirlas a texto plano separado por " | "
    list_cols = [col for col in df.columns if df[col].apply(lambda x: isinstance(x, list)).any()]

    for col in list_cols:
        df[col] = df[col].apply(lambda x: ' | '.join([str(i) for i in x]) if isinstance(x, list) else x)

    # 4. Guardar como CSV
    output_filename = 'recetas.csv'
    df.to_csv(output_filename, index=False, encoding='utf-8')

    print(f"¡Listo! Se ha creado el archivo '{output_filename}' con {len(df)} recetas.")

except Exception as e:
    print(f"Ocurrió un error: {e}")
```
