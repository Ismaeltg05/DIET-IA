# Añadir columna Calorías y Pasos

```python
# Importamos
from google.colab import drive
import pandas as pd
import ast

# Montamos Drive en nuestro espacio de trabajo
drive.mount('/content/drive')

# Cargamos los datos en un Dataframe
data = pd.read_csv('/content/drive/MyDrive/DIET-IA/dataset/DATASET FINAL/RAW_recipes.csv')

# Visualizamos las primeras filas para verificar alérgenos e ingredientes
print(data.head())
```

```python
# =========================
# 2. PARSEAR NUTRITION
# =========================
def parse_nutrition(x):
    if isinstance(x, list):
        return x
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return []
    return []

data["nutrition"] = data["nutrition"].apply(parse_nutrition)


# =========================
# 3. ASEGURAR 6 VALORES
# =========================
def fix_nutrition(n):
    if not isinstance(n, list):
        return [None]*6
    if len(n) < 6:
        return n + [None]*(6 - len(n))
    return n[:6]

data["nutrition"] = data["nutrition"].apply(fix_nutrition)


# =========================
# 4. SEPARAR COLUMNAS
# =========================
nutrition_cols = [
    "calories",
    "fat",
    "sugar",
    "sodium",
    "protein",
    "saturated_fat"
]

data[nutrition_cols] = pd.DataFrame(data["nutrition"].tolist(), index=data.index)


# =========================
# 5. LIMPIEZA FINAL
# =========================
data[nutrition_cols] = data[nutrition_cols].round(2)
```

```python
# Guardar nuevo dataset
output_path = '/content/drive/MyDrive/DIET-IA/dataset/DATASET FINAL/recipes_clean.csv'
data.to_csv(output_path, index=False)

print("Dataset limpio guardado en:", output_path)
```
